```python
import pandas as pd
import numpy as np
import pdfplumber
import re

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
import os

folder_path = "resumes"

files = os.listdir(folder_path) # List all files in the folder

for file in files:
    full_path = os.path.join(folder_path, file)
    print(full_path)
```

```python
from docx import Document
import pdfplumber
import os

def extract_text(file_path):
    if file_path.endswith(".docx"):
        doc = Document(file_path)
        return " ".join([p.text for p in doc.paragraphs]) # Extract text from .docx files

folder = "resumes"
data = []

for file in os.listdir(folder):
    path = os.path.join(folder, file)
    text = extract_text(path)
    data.append(text)

print(len(data), "resumes loaded")
print(data[0][:500]) # Print the first 500 characters of the first resume to verify extraction
```

```python
df = pd.DataFrame({
    "Resume_Text": data
}) # Create a DataFrame with the extracted resume texts 

df.head(10)
```

```python
import re

def clean_text(text):
    if not isinstance(text, str):
        return ""   # safely handle None / NaN
    text = text.lower()
    text = re.sub(r'\n+', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    return text # Clean the text by lowercasing, removing newlines, extra spaces, and special characters

df['Cleaned_Text'] = df['Resume_Text'].apply(clean_text)
df.head(10)# Check the cleaned text to ensure it's properly processed
```

```python
job_description = input("Enter the job description: ").strip() # Get the job description from the user and strip leading/trailing whitespace
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity # Import necessary libraries for generating embeddings and calculating similarity

model = SentenceTransformer('all-MiniLM-L6-v2') # Load the pre-trained SentenceTransformer model for generating embeddings

job_emb = model.encode(job_description) # Generate the embedding for the job description using the model
resume_embs = model.encode(df['Cleaned_Text'].tolist()) # Generate embeddings for all cleaned resume texts using the model
```

```python
scores = cosine_similarity(resume_embs, [job_emb]).flatten() # Calculate cosine similarity scores between each resume embedding and the job description embedding

df['NLP_Score'] = scores 
df_sorted = df.sort_values(by='NLP_Score', ascending=False) # Sort the DataFrame by the NLP similarity scores in descending order
print(df_sorted[['NLP_Score']].head(10)) # Print the top 10 resumes with their NLP similarity scores to the job description 
```

```python
required_skills = ["python", "machine learning", "sql"]

def skill_score(text):
    return sum(skill in text for skill in required_skills) / len(required_skills)

df['Skill_Score'] = df['Cleaned_Text'].apply(skill_score)# Calculate a skill score based on the presence of required skills in the cleaned resume text
df_sorted = df.sort_values(by=['NLP_Score', 'Skill_Score'], ascending=False) # Sort by NLP score and then by skill score
print(df_sorted[['NLP_Score', 'Skill_Score']].head(10))
```

```python
df['Final_Score'] = 0.7 * df['NLP_Score'] + 0.3 * df['Skill_Score'] # Combine NLP score and skill score into a final score with weights
df_sorted = df.sort_values(by='Final_Score', ascending=False) # Sort by the final combined score
print(df_sorted[['Final_Score', 'NLP_Score', 'Skill_Score']     ].head(10)) # Print the top 10 resumes with their final scores, NLP scores, and skill scores
```

```python
def explain(text):
    matched = [s for s in required_skills if s in text]
    missing = [s for s in required_skills if s not in text]
    return matched, missing
# Apply the explain function to each cleaned resume text to get matched and missing skills, and store them in new columns
df[['Matched', 'Missing']] = df['Cleaned_Text'].apply(lambda x: pd.Series(explain(x))) 
print(df[['Matched', 'Missing']].head(10))
```

```python
df['Decision'] = df['Final_Score'].apply(lambda x: "Selected" if x > 0.5 else "Rejected") # Make a simple decision based on the final score with a threshold
print(df[['Final_Score', 'Decision']].head(50)) # Print the final scores and decisions for the top 10 resumes    
```
